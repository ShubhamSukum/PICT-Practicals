```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("./Cleavland.csv",header=None)
df
```

```python
col=["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal","num"]
df.columns=col
```

```python
df
```

# histogram

```python
coli=["chol","fbs","restecg"]
df[coli].hist(bins=10)
plt.show()
```

# Dot plot

```python
sns.stripplot(x="age",y="cp",data=df,jitter=True)
```

```python
plt.plot(df["age"],df["cp"],"ro")
plt.show()
```

# Bar plot

```python
df["exang"].value_counts().plot(kind="bar")
plt.xlabel("x aahe re")
plt.ylabel("YYYY eee")
plt.show()
```

```python
#try
plt.figure(figsize=(30,30))
df["age"].value_counts().plot(kind="bar")
plt.show()
```

# Line Plot

```python
plt.figure(figsize=(20,20))
sns.lineplot(data=df,x="age",y="cp")
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.lineplot(data=df,x="thalach",y="slope")
plt.show()
```

# scatter plot

```python
plt.figure(figsize=(15,15))
sns.scatterplot(data=df,x="age",y="cp",hue="num")
plt.show()
```

# box

```python
sns.boxplot(data=df,x="age",y="cp")
```

# pie plot

```python
asd=df["cp"].value_counts()
plt.figure(figsize=(10,5))
plt.pie(asd,labels=asd.index,autopct="%1.1f%%")
plt.show()
```

```python
val=df["cp"].value_counts()
plt.figure(figsize=(10,10))
plt.pie(val,labels=val.index,autopct="%1.1f%%")
plt.show()
```
