```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
```

```python
request1=requests.get("https://www.flipkart.com/boat-airdopes-161-asap-charge-10mm-drivers-17-hours-playback-bluetooth-headset/p/itm8a7493150ae4a?pid=ACCG6DS7WDJHGWSH&lid=LSTACCG6DS7WDJHGWSHLM5ZET&marketplace=FLIPKART&store=0pm%2Ffcn%2F821%2Fa7x%2F2si&spotlightTagId=BestsellerId_0pm%2Ffcn%2F821%2Fa7x%2F2si&srno=b_1_2&otracker=browse&fm=organic&iid=012901ef-db86-4d59-ab71-d937e9c471c8.ACCG6DS7WDJHGWSH.SEARCH&ppt=sp&ppn=productListView&ssid=14s2jvo4vk0000001685171904801")
request1.content
```

```python
soup=bs(request1.text)
```

```python
reviews=soup.findAll('div',{'class':'t-ZTKy'})
all_reviews=[]
for review in reviews:
    reviewText=review.get_text().replace('READ MORE',"").strip()
    all_reviews.append(reviewText)
    print(reviewText+"\n\n")
```

```python
ratings=soup.findAll('div',{'class':'_1BLPMq'})
all_ratings=[]
for rating in ratings:
    all_ratings.append(rating.get_text())
all_ratings
```

```python
tags=soup.findAll('p',{'class':'_2-N8zT'})
all_tags=[]
for tag in tags:
    all_tags.append(tag.get_text())
all_tags
```

```python
cus_names=soup.findAll('p',{'class':'_2sc7ZR _2V5EHH'})
all_cus_names=[]
for cus_name in cus_names:
    all_cus_names.append(cus_name.get_text())
all_cus_names
```

```python
data={
    "names":all_cus_names,
    "ratings":all_ratings,
    "tags":all_tags,
    "reviews":all_reviews,
}
data
```

```python
pd.set_option('display.max_colwidth',None)
df=pd.DataFrame(data)
df
```

```python
df.to_csv('PS31.csv')
```
