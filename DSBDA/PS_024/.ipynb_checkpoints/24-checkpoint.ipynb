```python
import pandas as pd
```

```python
DataFrame=pd.read_csv('forestfires.csv')
DataFrame
```

# Subset
### Required to done by Classes here but not implemented in that way 

```python
Subset1=DataFrame[DataFrame['area']>=0.50]
Subset1
```

```python
Subset2=DataFrame[DataFrame['area']<0.50]
Subset2
```

# Merge

```python
Merged=pd.concat([Subset1,Subset2])
Merged
```

# Sort

```python
Sorted1=DataFrame.sort_values('temp')
Sorted1
```

```python
Sorted2=DataFrame.sort_values('wind')
Sorted2
```

```python
Sorted2=DataFrame.sort_values('area')
Sorted2
```

# Melt

```python
Melt=pd.melt(DataFrame,id_vars=['month','day','temp'],var_name='Metric',value_name='value')
Melt
```

# Casting

```python
Casting=pd.pivot_table(DataFrame,index=['month','day','temp'],values='area')
Casting
```
